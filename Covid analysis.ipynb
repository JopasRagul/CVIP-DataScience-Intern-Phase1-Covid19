```python
# bar plot for state and death count values by 'Death'
# define x and y
x=dataset[['State/UTs','Deaths']].sort_values(by='Deaths',ascending=False)['State/UTs'][:10].values # first 10 values

y=dataset[['State/UTs','Deaths']].sort_values(by='Deaths',ascending=False)['Deaths'][:10].values # first 10 values

# dataframe
df=pd.DataFrame({'State/UTs':x,
                'Deaths':y})
#bar plot
fig=px.bar(df,
          x='State/UTs',
          y='Deaths',
          color='State/UTs',
          title='State/UTs Vs No. of Deaths')
fig.show() # show the figure
```

```python
# bar plot for state and total case by death ratio
px.bar(dataset, x="State/UTs", y="Total Cases", color="Death Ratio", title="Total Cases as per each State : ")
```

```python
fig = px.choropleth(
    dataset,
    geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
    featureidkey='properties.ST_NM',
    locations='State/UTs',
    color='Total Cases',
    color_continuous_scale='reds'
)

fig.update_geos(fitbounds="locations", visible=True)

fig.show()
```

```python
#State with  more death ratio in India
# named as dataset1
dataset1 = dataset.sort_values(by='Death Ratio', ascending=False).head()
states = dataset1['State/UTs']
ratio = dataset1['Death Ratio']
plt.barh(states, ratio)
plt.xlabel('Death Ratio')
plt.ylabel('State')
plt.title('State with  more death ratio in India')
plt.show()
```
